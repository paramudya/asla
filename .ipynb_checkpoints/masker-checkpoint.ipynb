```python
f='list_cif.xlsx'
d=pre_masked[f].copy()

###group modification
for old, new in map_grup_idx.items(): # takes long as it loops lots dict items
    d['row_key'] = d['row_key'].str.replace(old, new)   
    d['grup'] = d['grup'].str.replace(old, new)
    # d['nama_entitas'] = d['grup']+d['nama_entitas'].str[-5:]+d['nama_entitas'].str[d['cif_key'].astype(str).str[4]:]+d['nama_entitas'].str[:4].str.replace(old, new)

##entity modification
def if_import(mapping,x,chance_reverse):
    if int(x%chance_reverse)==1: return '' #1/chance_reverse of not returning
    else: return mapping[x] #chatgpt_generated_indo_industry_word_157
entity_complex_1=d['nama_entitas'].astype('str').apply(lambda x: str(x)[int(str(len(x)*61668)[-2]):int(str(len(x)*61668)[-2])+3])
tmp_str_modif_1=(d['nama_entitas'].str[-23:-18].str[::-1]+entity_complex_1+d['nama_entitas'].str[3:7].str[::-1]).apply(lambda x: re.sub(' +', '',str(x)))

tmp_str_modif_1=pd.DataFrame(tmp_str_modif_1)
tmp_str_modif_1[tmp_str_modif_1['nama_entitas']=='']='tuku simprug'
tmp_str_modif_1=tmp_str_modif_1['nama_entitas']
tmp_to_map=tmp_str_modif_1.apply(lambda x:int(str(int(ord(x[-1])/ord(x[min(-1,-int(len(x)/1.2))])*ord(x[max(0,int(len(x)/1.9))])*ord(x[0])*len(x) ))[max(-len(x),-5):]))

nama_entitas_tmp=(tmp_to_map.apply(lambda x:if_import(chatgpt_andme_generated_indo_industry_word_163,(x)%157,2))+' ' +\
tmp_to_map.apply(lambda x:if_import(map_kata_idx_1305,(x*133)%1305,4))+' '+ \
tmp_str_modif_1.str[:5]+' '+ \
tmp_to_map.apply(lambda x:if_import(chatgpt_andme_generated_indo_industry_word_163,int(x*61668)%157,4))+' ' +\
tmp_to_map.apply(lambda x:if_import(map_kata_idx_1305,x%1305,3)) \
).str.title().apply(lambda x: re.sub(' +', ' ',str(x)).strip())

map_entity_bni={}
placeholder_nama_entitas_loop=[]
ct_gagrup,ct_grup=0,0
for i,row in pre_masked['list_cif.xlsx'].iterrows():
    if str(row['grup']) in str(row['nama_entitas']): 
        new_name=(row['grup']+' '+nama_entitas_tmp[i]).title()
        ct_grup+=1
    else: 
        ct_gagrup+=1
        new_name=nama_entitas_tmp[i].title()
    placeholder_nama_entitas_loop.append(new_name)
    map_entity_bni[str(row['nama_entitas'])]=new_name
d['nama_entitas']=placeholder_nama_entitas_loop
print(ct_gagrup,'grup:',ct_grup)
###divisi modification
for old, new in map_divisi.items():
    d['row_key'] = d['row_key'].str.replace(old, new)   
    d['divisi'] = d['divisi'].str.replace(old, new)

###cif modification
d['cif_key']=d['cif_key']+int(str(d['cif_key']%10000001-d['cif_key']%10000000+d['cif_key']%2000)[-2:]) #consistent (always the same eevery run) encrypting, addition
cifs_str=d['cif_key'].astype(str) #str replace
cif_2_6=(cifs_str.apply(lambda x: x if x.strip() in ['nan',''] else str(int(x[1:max(min(6,int(int(x[:4])%int(x[:2]))),4)] )*16056+61668)).str[-5:])
d['cif_key']=d['cif_key'].astype(str).str[-13:-11]+cif_2_6+d['cif_key'].astype(str).str[-5:]

####nominal modification
nominal_cols=['posisi_saldo_eom','avg_saldo_ytd','baki_debet_eom','avg_baki_debet_ytd','maksimum_kredit'] #utilisasi kredit,ldr_eom,ldr_ytd ga dimodif karena anggapannya dependen sama yang lain secara proporsi 

pengali=1+(d['grup'].apply(lambda x:(ord(x[0].lower())%68-46)/100)) #slightly distort the total number (lower), -17 sampai 8
for col in nominal_cols: d[col]=pengali*d[col]
d['maksimum_kredit']=d['maksimum_kredit'].astype(str).str.split('.').apply(lambda x:x[0]).value_counts().head(10)

masked[f]=d.copy()
```

```python
f='value_chain.xlsx'
d=pre_masked[f].copy()

cols_otherbank=[
    'potensi_cash_in_other_bank_top5', #max 10
    'potensi_cash_out_other_bank_top5',  #max 10
]
cols_bni=['potensi_cash_in_bni_top10','potensi_cash_out_bni_top10']
cols_bank=[
    'potensi_cash_in_other_bank_top3_bank',
    'potensi_cash_out_other_bank_top3_bank'
]

#out
all_unique_out=set([xs for i in range(len(cols_otherbank)) for x in d[cols_otherbank[i]].str.split('~') \
    .apply(lambda x: x[1::2] if type(x)==list else [])  for xs in x]) #1738. all uniqque companies from other bank
out_4worded_941=dict(enumerate(set([ws for w in all_unique_out if len(w.split())>3 for ws in w.split()]))) #941 words
mask_out_word_sources={
    0:chatgpt_andme_generated_indo_industry_word_163,
    1:map_kata_idx_1305,
    2:out_4worded_941
    } 
map_out= {
    out: mask_out_word_sources[(ord(out[0])*ord(out[-1]))%3][(word_no+ord(out[0]))*ord(out[-1])*len(out)%len(mask_out_word_sources[(ord(out[0])*ord(out[-1]))%3])] \
        for out in all_unique_out \
            for word_no in range(len(out.split()))    } #1738. 
for old, new in map_out.items(): # takes long as it loops lots dict items
    for c in cols_otherbank:
        d[c] = d[c].str.replace(old, new)

#in
map_in_diluar_yg_ada= {
    out: mask_out_word_sources[(ord(out[0])+ord(out[-1]))%3][(word_no+ord(out[0]))*ord(out[-1])*len(out)%len(mask_out_word_sources[(ord(out[0])+ord(out[-1]))%3])] \
        for out in all_unique_out \
            for word_no in range(len(out.split()))    } #1738. 

for old, new in map_entity_bni.items(): 
    for c in cols_bni:
        d[c] = d[c].str.replace(old, new) # gaada ujung2nya

for old, new in map_in_diluar_yg_ada.items(): # takes long as it loops lots dict items
    for c in cols_bni:
        d[c] = d[c].str.replace(old, new)  

#bank, should be the most simplest
bank_count=pd.DataFrame(pd.Series([xs for i in range(len(cols_bank)) for x in d[cols_bank[i]].str.split('~') \
    .apply(lambda x: x[1::2] if type(x)==list else [])  for xs in x]).value_counts())
bank_count=dict(bank_count['count'].apply(lambda x: int(max(1, int(x *ord(str(x)[min(3,len(str(x))-1)]) + (int(str(x)[min(1,len(str(x))-1) ])- len(str(x).replace('1',''))) )%40 + x/3 ) )))
lst_bank_weights=[item for sublist in [[key]*value for key,value in bank_count.items()] for item in sublist]
map_bank={list(set(bank_count.keys()))[i]:lst_bank_weights[(i**2)%len(lst_bank_weights)] for i in range(len(set(bank_count.keys())))}

for old, new in map_bank.items(): # takes long as it loops lots dict items
    for c in cols_bank:
        d[c] = d[c].str.replace(old, new)  


masked[f]=d.copy()
```
